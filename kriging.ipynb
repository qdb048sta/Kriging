```python
import os
from datetime import datetime
path="D:\\User_Data\\Desktop\\kriging\\"
os.chdir(path)
import pandas as pd
import numpy as np
def date_to_delta(x):#transform numerate date to datetime
    x=str(x)
    x=x[0:8]
    time=datetime.strptime(x,"%Y%m%d")
    return time
import numpy as np
import pykrige.kriging_tools as kt
from pykrige.ok import OrdinaryKriging
import matplotlib.pyplot as plt
import logging
FORMAT = '%(asctime)s %(levelname)s: %(message)s'
logging.basicConfig(level=logging.DEBUG, format=FORMAT,filename='mylog.log', filemode='a')
import warnings
warnings.filterwarnings('ignore')
def delta_to_date(x):
    strd=x[0:10]
    strlist=strd.split("-")
    return strlist[0]+strlist[1]+strlist[2]
    
    
```

```python
df=pd.read_csv(path+"kriging.csv")
pollution_set=["PM2.5","PM10","SO2","NO2"]
tw_grid=pd.read_csv(path+"tw_grid_3km.csv")
gridx=tw_grid["x"]
gridy=tw_grid["y"]
for p in pollution_set:
    for v in ["avg","max"]:
        df=pd.read_csv("E:\\poll\\SECOND_STEP\\original_data_"+p+"_"+v+".csv")
        df=df[df[v]>0]
        sub_df_1=df[["date","x","y",v]]
        sub_df_1["date_delta"]=sub_df_1["date"].apply(date_to_delta)
        sub_df_1=sub_df_1.sort_values("date_delta",ascending=False)
        sub_df_1=sub_df_1[sub_df_1["date_delta"]>=datetime(1998,1,1)]
        date_delta=set(sub_df_1["date_delta"])
        print(sub_df_1.head(5))
        sub_df_or=pd.DataFrame([["19600101",-1,-1,-1,-1,p]],columns=["date","x","y","predicted_value","predicted_std","pol"])
        for d in date_delta:
            sub_df_2=sub_df_1[sub_df_1["date_delta"]==d]
            sub_df_2=sub_df_2.dropna(how="any")
            if len(sub_df_2)>0:
                data=np.array(sub_df_2)
                try:
                    OK = OrdinaryKriging(
                        data[:, 1],
                        data[:, 2],
                        data[:, 3],
                        variogram_model="linear",
                        verbose=False,
                        enable_plotting=False)
                    z, ss = OK.execute("points",gridx,gridy)
                    sub_df_3=pd.DataFrame()
                    sub_df_3["x"]=gridx
                    sub_df_3["y"]=gridy
                    sub_df_3["predicted_value"]=z
                    sub_df_3["predicted_std"]=ss
                    sub_df_3["date"]=np.array([str(d)]*len(sub_df_3))
                    sub_df_3["date"]=np.array([str(d)]*len(sub_df_3))
                    sub_df_3["pol"]=np.array([p]*len(sub_df_3))
                    sub_df_or=pd.concat([sub_df_or,sub_df_3])
                except Exception as e:
                    print(str(e))
                    print(p+"_"+v+"_"+str(d))
            else:
                    logging.debug(str(d)+" is empty")
            logging.debug(print(sub_df_or.shape))
        sub_df_or=sub_df_or[sub_df_or["x"]>=0]
        try:
            sub_df_or["date"]=sub_df_or["date"].apply(delta_to_date)
            sub_df_or.to_csv(path+p+"_"+v+"_"+"kriging.csv")
        except:
            sub_df_or.to_csv(path+p+"_"+v+"_"+"kriging.csv")
                

                    

                
                
            
            
            
            
    print(datetime.now())
            
            
            
            
            
            
            
        
        
        
        
    
    
```
